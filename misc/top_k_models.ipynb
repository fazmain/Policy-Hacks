```python
import pickle

with open("huggingface_model_cards.pkl", "rb") as f:
    model_cards = pickle.load(f)

total_models = len(model_cards)
print("Total models:", total_models)
```

```python
import json

with open('huggingface_model_cards.json', 'r') as f:
    cards_json = json.load(f)

model_ids = [card['model_id'] for card in cards_json]
print(model_ids)
```

```python
with open("model_ids.md", "w") as f:
    f.write("# Model IDs\n\n")
    for model_id in model_ids:
        f.write(f"- `{model_id}`\n")
print("Exported model_ids to model_ids.md")
```
